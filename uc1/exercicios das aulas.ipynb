```python
def calcular_media(lista_notas):
    soma = sum(lista_notas)
    media = soma / len(lista_notas)
    return media
```

```python
alunos = {'Lucas': [7,8,5,7],
          'Karol': [10,5,7,6]}

resultado = calcular_media(alunos['Karol'])
print(resultado)
```

```python
def tratar_nº_int(mensagem):
    while True:
        try:
            valor = float(input(mensagem))
            return
        except ValueError:
            print("Digite um numero valido!")

idade = tratar_nº_int("Digite sua idade: ")
print(f"Idade registrada: {idade}")

```

```python
def tratar_nº_float(mensagem):
    while True:
        try:
            valor = float(input(mensagem))
            return valor
        except ValueError:
            print("Digite um numero valido!")

def calcular_area(a,b):
    area = a * b
    return area

base = tratar_nº_float("Digite a base do empreendimento: ")
altura = tratar_nº_float("Digite a altura do empreendimento? ")

are = calcular_are(base, altura)
print(f'A área é de {area} m2.')


```
